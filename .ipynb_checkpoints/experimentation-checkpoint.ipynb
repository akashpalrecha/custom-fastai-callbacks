## Setting up model for experimentation

```python
PATH = Path('/Users/akash/Desktop/AI/Deep learning/GitHub Uploads/Donkeys vs Horses/data')
def get_data(bs=16, size=224):
    data = (ImageItemList.from_folder(path=PATH)
            .filter_by_folder(include=['Donkey', 'Horse'])
            .random_split_by_pct(0.2)
            .label_from_folder()
            .transform(get_transforms(max_warp=0), size=size)
            .databunch(bs=bs)
            .normalize(imagenet_stats))
    return data
```

```python
data = get_data()
learn = create_cnn(data, models.resnet18)
```

```python
class SaveAfterNIterations(LearnerCallback):
    """
    Save Model after N Batches
    This is useful when training models that take hours for just 1 epoch
    """
    def __init__(self, learn:Learner, num_batches:int = 100, save_name=None):
        super().__init__(learn)
        self.num_batches = num_batches
        self.save_name = save_name
        if save_name is None:
            self.save_name = f'saved_every_{self.num_batches}_epochs'
    
    def on_batch_end(self, num_batch, iteration, **kwargs:Any) ->None:
        if num_batch % self.num_batches == 0:
            self.learn.save(name=self.save_name)
            print(f"Model saved as {self.save_name}\n Iteration : {iteration} | Batch : {num_batch}")
```

```python
saver_callback = partial(SaveAfterNBatches, num_batches=10)
```

```python
data = get_data(4, 224)
```

```python
learn = create_cnn(data, models.resnet18, callback_fns = [saver_callback])
```

```python
learn.fit_one_cycle(1)
```
